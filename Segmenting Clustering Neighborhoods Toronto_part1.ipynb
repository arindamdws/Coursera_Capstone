```python
sources=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
print("data downloaded")
```

## Get the data from witipedia using request method

```python
soup=BeautifulSoup(sources,'lxml')
df2=pd.DataFrame(['7']) ## This will be used to concatenate the rows during the loop introduced in the code below
print(df2)
```

## Instantiate the dataframe

```python
table = soup.table
table_rows = table.find_all('tr')
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    df1 = pd.DataFrame(data=row,dtype='str').transpose()
    df2 = pd.concat([df1,df2])
    print(df1)
```

## Print the data frame in for loop method

```python
df2.columns=['PostalCode', 'Borough','Neighborhood']
#print(df2)
df2 = df2.set_index("Borough")
df2 = df2.drop("Not assigned", axis=0) # Delete all rows with label "Not assigned"

df2=df2.groupby(['PostalCode', 'Borough'])['Neighborhood'].\
    agg({ 
         'Neighborhood': lambda x: ','.join(x)
        }
       )

df2 = df2.reset_index()
df2
```

#### Delete all records whose Borough were having value "Not assigned" 
#### Concatenate the values in the neighborhood column that has same Postal code
#### Display the data frame

```python
df2.replace({  '\n' : ''}, regex=True, inplace=True)
df2.shape
```
